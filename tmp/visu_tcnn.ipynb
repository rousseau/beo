```python
import matplotlib.pyplot as plt
import nibabel
from scipy import ndimage
from os.path import expanduser
home = expanduser("~")
```

```python
images_gt = []
data_gt = []

n=15
for i in range(n):
  images_gt.append(nibabel.load(home+'/Sync-Exp/t2-t'+str(21+i)+'.00_128.nii.gz'))
  data_gt.append(images_gt[-1].get_fdata())


fig=plt.subplots(nrows=1,ncols=15,figsize=(30,20))
for i in range(n):
  plt.subplot(1, n, int(i+1))
  plt.imshow(ndimage.rotate(data_gt[i][:,:,64],90),cmap='gray')
plt.show()

#Reconstructions using all images and 15 encodings
data_15enc_15im = nibabel.load(home+'/Sync-Exp/pred_15enc_15im.nii.gz').get_fdata()

fig=plt.subplots(nrows=1,ncols=15,figsize=(30,20))
for i in range(n):
  plt.subplot(1, n, int(i+1))
  plt.imshow(ndimage.rotate(data_15enc_15im[:,:,64,i],90),cmap='gray')
plt.show()

#Reconstructions using all images and 2 encodings
data_2enc_15im = nibabel.load(home+'/Sync-Exp/pred_2enc_15im.nii.gz').get_fdata()

fig=plt.subplots(nrows=1,ncols=15,figsize=(30,20))
for i in range(n):
  plt.subplot(1, n, int(i+1))
  plt.imshow(ndimage.rotate(data_2enc_15im[:,:,64,i],90),cmap='gray')
plt.show()

fig=plt.subplots(nrows=1,ncols=15,figsize=(30,20))
for i in range(n):
  plt.subplot(1, n, int(i+1))
  plt.imshow(ndimage.rotate(data_15enc_15im[:,:,64,i]-data_2enc_15im[:,:,64,i],90),cmap='gray')
plt.show()

#Reconstructions using 3 images and 2 encodings
data_2enc_3im = nibabel.load(home+'/Sync-Exp/pred_2enc_3im.nii.gz').get_fdata()

fig=plt.subplots(nrows=1,ncols=15,figsize=(30,20))
for i in range(n):
  plt.subplot(1, n, int(i+1))
  plt.imshow(ndimage.rotate(data_2enc_3im[:,:,64,i],90),cmap='gray')
plt.show()

fig=plt.subplots(nrows=1,ncols=15,figsize=(30,20))
for i in range(n):
  plt.subplot(1, n, int(i+1))
  plt.imshow(ndimage.rotate(data_15enc_15im[:,:,64,i]-data_2enc_3im[:,:,64,i],90),cmap='gray')
plt.show()

#Reconstructions using 2 images and 2 encodings
data_2enc_2im = nibabel.load(home+'/Sync-Exp/pred_2enc_2im.nii.gz').get_fdata()

fig=plt.subplots(nrows=1,ncols=15,figsize=(30,20))
for i in range(n):
  plt.subplot(1, n, int(i+1))
  plt.imshow(ndimage.rotate(data_2enc_2im[:,:,64,i],90),cmap='gray')
plt.show()

fig=plt.subplots(nrows=1,ncols=15,figsize=(30,20))
for i in range(n):
  plt.subplot(1, n, int(i+1))
  plt.imshow(ndimage.rotate(data_15enc_15im[:,:,64,i]-data_2enc_2im[:,:,64,i],90),cmap='gray')
plt.show()
```
